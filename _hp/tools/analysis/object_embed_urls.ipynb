```python
import pandas as pd
```

```python
df = pd.read_csv("embed_object.csv", index_col=0)
```

```python
df.head()
```

```python
df.groupby(["browser", "element"]).describe()
```

```python
df.agg(["unique"])
```

```python
d = df.pivot(index=['file_ending'], columns=['browser', 'element'], values='executes')
display(d)
```

```python

# ALl the same
print("All the same:")
display(d[d.nunique(axis=1) == 1])

# Different outcomes
print("Not all the same:")
display(d[d.nunique(axis=1) != 1])

print("All forbidden:")
display(d[d.sum(axis=1) == 0])
print("All allowed:")
display(d[d.sum(axis=1) == 6])
print("One allowed:")
display(d[d.sum(axis=1) == 1])
print("Two allowed:")
display(d[d.sum(axis=1) == 2])
print("Three allowed:")
display(d[d.sum(axis=1) == 3])
print("Four allowed:")
display(d[d.sum(axis=1) == 4])
print("Five allowed:")
display(d[d.sum(axis=1) == 5])
```

```python
with pd.option_context("display.max_rows", 111):
    display(d[d.sum(axis=1) == 4])
```

```python
d.to_csv("embed-object-res.csv")
```
